```python
from collections import defaultdict, deque, OrderedDict

# read graph data file
lines = []
with open('GraphData.txt') as f:
    lines = f.readlines()
lines = [x.strip() for x in lines]

    
#prepare graph
graph = defaultdict(list)
nodes = []
for line in lines:
    n1 = line.split()[0].replace(',','').replace('(', '')
    n2 = line.split()[1].replace(')','')
    graph[n1].append([n2, abs(int(n1)- int(n2))])
```

```python
n_vertex = len(graph.keys())
```

```python
def find( parent, i):
    if parent[i] == i:
        return i
    return find(parent, parent[i])
```

```python
def union(parent, rank, x, y):
    xroot = find(parent, x)
    yroot = find(parent, y)
    if rank[xroot] < rank[yroot]:
        parent[xroot] = yroot
    elif rank[xroot] > rank[yroot]:
        parent[yroot] = xroot
    else :
        parent[yroot] = xroot
        rank[xroot] += 1
```

```python
result =[]
i,e = 0,0
graph = OrderedDict(sorted(graph.items(), key=lambda item: float(item[0])))
parent = [] ; rank = []
for node in range(n_vertex):
    parent.append(node)
    rank.append(0) 
while e < n_vertex -1 :
    for key in graph.keys():
        for item in graph[key]:
            u,v,w =  int(key), int(item[0]), int(item[1])

            i = i + 1
            x = find(parent, u)
            y = find(parent ,v)
            if x != y:
                e = e + 1  
                result.append([u,v,w])
                union(parent, rank, x, y) 
```
