```python
from collections import defaultdict


graph = defaultdict(list)

with open('GraphData.txt') as f:
    nodes = f.readline().replace('(', '').replace(')', '') #
    n1 = nodes.split()[0].replace(',','')
    n2 = nodes.split()[1]
    graph[n1].append(n2)
```

```python
lines = []
with open('GraphData.txt') as f:
    lines = f.readlines()
```

```python
lines = [x.strip() for x in lines]
```

```python
graph = defaultdict(list)
for line in lines:
    n1 = line.split()[0].replace(',','')
    n2 = line.split()[1]
    graph[n1].append(n2)
```
